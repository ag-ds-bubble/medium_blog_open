The basic idea here is to find an set of embeddings from the given dataset using neural nets so that those embeddings can be used as an input feature to the models being trained so far.

Why didnt we look at Neural Net earlier?
- Very small dataset, we deal with this using bootstrapping and adjusting the target variable values in the range of its standard deviation. (Will clear this over a call)


So idea here is to -

    - Deal with the data
    1.1) BootStrap the dataset
    1.2) Replace the target variable with a new values which would be a deviation from the actuals within the range of +STD and -STD
    
    - Neural Net Methoddology
    2.1) Simple Neural net training on bootstrapped data
    2.2) Auto-Encoder based feature generation
    2.3) (2.2) in rolling methodology

# Helper Functions

```python
class HelperFunctions:
    def __init__(self):
        print('Initialising Helper Function Class...')
        
    def create_bootstrappeddata(self, df, targetcol, newsize, _features):
        
        df = df[_features].dropna().reset_index().drop('index', axis=1)
        _std = df[targetcol].dropna().std()*0.5
        bootstrappeddata = pd.DataFrame()
        
        for _ in tqdm(range(newsize)):
            temp = df.sample(1)
            targetval = temp[targetcol].values[0]
            possibletargets = np.linspace(targetval-_std, targetval+_std, 20).tolist()
            temp[targetcol] = random.sample(possibletargets, k=1)
            bootstrappeddata = pd.concat([bootstrappeddata, temp])
            
        return bootstrappeddata
    
    def calc_mape(self, yTrue, yPred):
                
        _diff    = yTrue-yPred
        _absDiff = np.abs(_diff)
        _absPct  = _absDiff/yTrue
        _mape    = np.mean(_absPct)*100
        
        return _mape
    

helperFunctionHandler =  HelperFunctions()
```

```python
class ModelArchitecture:
    def __init__(self):
        print('Initialising the model architecture class..')
        model = None
        
    def model_arch_1(self, inp_shape):
        model = Sequential()
        # L1
        model.add(Dense(50, activation='relu', kernel_initializer='he_uniform', input_dim= inp_shape[1]))
        model.add(BatchNormalization())
        # L2
        model.add(Dense(30, activation='relu', kernel_initializer='he_uniform'))
        model.add(BatchNormalization())
        # L3
        model.add(Dense(30, activation='relu', kernel_initializer='he_uniform'))
        model.add(BatchNormalization())
        # L4
        model.add(Dense(30, activation='relu', kernel_initializer='he_uniform'))
        model.add(BatchNormalization())
        # L5
        model.add(Dense(20, activation='relu', kernel_initializer='he_uniform'))
        model.add(BatchNormalization())
        # L6
        model.add(Dense(20, activation='relu', kernel_initializer='he_uniform'))
        model.add(BatchNormalization())
        # Output Layer
        model.add(Dense(1, activation='sigmoid'))
        opt = SGD(lr=0.0001, momentum=0.9,nesterov=True)
        # opt = Adagrad()
        model.compile(loss='mse', optimizer=opt, metrics=['mae', 'mse'])
        model = model
        display(model.summary())
        
        return model

modelArchHandler = ModelArchitecture()
```

```python
_carmodeldata_Train_X = bootstrapped_data[_features].drop(target_variable, axis=1)[:-1000].copy()
_carmodeldata_Train_y = bootstrapped_data[target_variable][:-1000].copy()

_carmodeldata_Train_X_dev = bootstrapped_data[_features].drop(target_variable, axis=1)[-1000:].copy()
_carmodeldata_Train_y_dev = bootstrapped_data[target_variable][-1000:].copy()

_carmodeldata_Test_X = _carmodeldata_Test[_features].drop(target_variable, axis=1).copy()
_carmodeldata_Test_y = _carmodeldata_Test[target_variable].copy()

_carmodeldata_Forecast_X = _carmodeldata_Forecast.copy()
```

## Train Model

```python
training_logger = CSVLogger('training.log')
nn_model = modelArchHandler.model_arch_1(inp_shape = _carmodeldata_Train_X.shape)
nn_model_traincallback = nn_model.fit(x=_carmodeldata_Train_X, y=_carmodeldata_Train_y, 
                                      batch_size=TRAINING_BATCHSIZE,
                                      validation_data=(_carmodeldata_Train_X_dev, _carmodeldata_Train_y_dev), 
                                      epochs=TRAINING_EPOCHS, verbose=1, 
                                      callbacks=[training_logger])
```

# Random Testing Space

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from mpl_toolkits.mplot3d.axes3d import get_test_data



var_x = 'EngineeredFeature@SQRT.AGE_FE'
var_y = 'EngineeredFeature@TOTAL_INCENTIVES_FELag5'
var_z = 'volume@Nissan_Leaf-Volume'

pltData = _modellingdata[[var_x, var_y, var_z]].dropna()
_mode = 'ot'

class TestingClass:
    
    def __init__(self):
        print('Initialising Testing Class...')
    
    def create_bootstrappeddata(self, df, targetcol, newsize, _features, std_scaling=0.1, mode='ot'):

        df = df[_features].dropna().reset_index().drop('index', axis=1)

        bootstrappeddata = pd.DataFrame()

        for _ in tqdm(range(newsize)):
            temp = df.sample(1)
            if mode == 'ot':
                topertub = [targetcol]
            elif mode == 'of':
                topertub = [k for k in _features if k!=targetcol]
            elif mode == 'all':
                topertub = _features

            for eachFeat in topertub:

                _std = df[eachFeat].dropna().std()*std_scaling
                origval = temp[eachFeat].values[0]
                possiblevals = np.linspace(origval-_std, origval+_std, 20).tolist()
                temp[eachFeat] = random.sample(possiblevals, k=1)
                bootstrappeddata = pd.concat([bootstrappeddata, temp])

        return bootstrappeddata
    
    def plotData(self, pltdata, pltbootdata, mode):
        
        fig = plt.figure(figsize=plt.figaspect(0.5))

        ax = fig.add_subplot(1, 2, 1, projection='3d')
        X = pltdata[var_x]
        Y = pltdata[var_y]
        Z = pltdata[var_z]

        ax.scatter(X, Y, Z)
        ax.set_title('Actual')


        # #---- Second subplot
        ax = fig.add_subplot(1, 2, 2, projection='3d')
        X = pltbootdata[var_x]
        Y = pltbootdata[var_y]
        Z = pltbootdata[var_z]
        ax.scatter(X, Y, Z)
        ax.set_title(f'BootStrapped - Mode : {mode}')
        plt.close()
        return fig

testingClassHandler = TestingClass()

pltDataBootStrapped = testingClassHandler.create_bootstrappeddata(df = pltData, 
                                                                  targetcol = target_variable,
                                                                  newsize = 1000,
                                                                  _features = pltData.columns.tolist(),
                                                                  std_scaling = 0.05,
                                                                  mode=_mode)
testingClassHandler.plotData(pltdata = pltData.copy(),
                             pltbootdata = pltDataBootStrapped.copy(),
                             mode = _mode)

```
